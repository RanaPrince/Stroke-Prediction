```python
df.plot(kind='scatter',y= 'bmi',x='age')
plt.title('Scatter plot')
plt.show()
```

```python
# 2-D Scatter plot with color-coding for each status.
sns.set_style("whitegrid");
sns.FacetGrid(df, hue="stroke", height = 7) \
   .map(plt.scatter,  "age", "bmi") \
   .add_legend();
plt.title('ScatterPlot-Age vs bmi')
plt.show();
```

```python
sns.FacetGrid(df,hue="stroke",height = 5)\
    .map(sns.distplot,"bmi")\
    .add_legend()
plt.title('Nodes plot')
plt.show();
```
